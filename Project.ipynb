### Encoding the Target Class Variable

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['class'] = label_encoder.fit_transform(df['class'])
```

#### Undersampling

There is a large class imbalance, so undersampling maybe needed.

```python
X=df.drop('class', axis=1)
y=df['class']
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
undersampler = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = undersampler.fit_resample(X, y)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x=y_resampled, palette='viridis')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Count Plot of Categories after Under Sampling')
```

```python
from sklearn.feature_selection import GenericUnivariateSelect, f_classif, mutual_info_classif
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
X_train_log = np.log1p(X_train)
X_test_log = np.log1p(X_test)
X_train_log=X_train_log.fillna('0')
X_train_log.info()
```

```python
selector = GenericUnivariateSelect(score_func=f_classif, mode='k_best', param=8)
X_selected = selector.fit_transform(X_train_log, y_train)
selected_features = selector.get_feature_names_out()
print("Selected features", selected_features)
```

```python
selector = GenericUnivariateSelect(score_func=mutual_info_classif, mode='k_best', param=8)
X_selected = selector.fit_transform(X_train_log, y_train)
selected_features = selector.get_feature_names_out()
print("Selected features", selected_features)
```

```python
X_train_log=X_train_log.drop(['objid', 'ra', 'dec', 'run', 'rerun', 'camcol', 'field', 'mjd','fiberid'], axis=1)
X_test_log=X_test_log.drop(['objid', 'ra', 'dec', 'run', 'rerun', 'camcol', 'field', 'mjd','fiberid'], axis=1)
```

```python
X_train_log.head()
```

### Model Testing

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, ExtraTreesClassifier, BaggingClassifier
from sklearn.neural_network import MLPClassifier

from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
ml_models= [
    LogisticRegression(),
    SVC(),
    RandomForestClassifier(),
    KNeighborsClassifier(),
    GaussianNB(),
    DecisionTreeClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    ExtraTreesClassifier(),
    BaggingClassifier(),
    MLPClassifier(),
]
model_names=['LogisticRegression', 'SVC', 'RandomForestClassifier', 'KNeighborsClassifier', 'GaussianNB',
    'DecisionTreeClassifier', 'AdaBoostClassifier', 'GradientBoostingClassifier', 'ExtraTreesClassifier',
    'BaggingClassifier', 'MLPClassifier']
```

```python
accuracies=[]
for model in ml_models:
    pipe= make_pipeline(model)
    pipe.fit(X_train_log, y_train)
    pred=pipe.predict(X_test_log)
    accuracy=accuracy_score(y_test, pred)
    accuracies.append(accuracy)
pipe_out={'ML Model': model_names, 'Accuracies':accuracies}
model_df=pd.DataFrame(pipe_out)
model_df
```

```python
model_df=model_df.sort_values(by='Accuracies', ascending=False)
model_df=model_df.reset_index(drop=True)
```

```python
model_df
```

```python
from sklearn.model_selection import GridSearchCV
random_forest=RandomForestClassifier()
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_search = GridSearchCV(estimator=random_forest, param_grid=param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train_log, y_train)
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print("Best Set of Parameters:", best_params)
print("Best Score:", best_score)
best_rf = grid_search.best_estimator_
test_accuracy = best_rf.score(X_test_log, y_test)
print("Test Accuracy with Best Parameters:", test_accuracy)
```

```python
y_preds = grid_search.predict(X_test_log)

confmatrix = confusion_matrix(y_test, y_preds)
print("Confusion Matrix:")
print(confmatrix)
```

```python
classreport = classification_report(y_test, y_preds)
print("Classification Report:")
print(classreport)
```
